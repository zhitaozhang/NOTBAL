```python
from astropy.io import fits
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
noBAL_file = fits.open('DR12_noBAL.fits')
```

```python
noBAL_file.info()
```

```python
from astropy.table import Table
NOTBAL = Table.read(noBAL_file,1)
print(NOTBAL.columns)
```

```python
#data1 = noBAL_file.header['flux']

data = noBAL_file[1].data
wave = data[0]['wave']
flux = data[0]['flux']
error= data[0]['sig']
```

```python
plt.figure(num=3, figsize=(18, 3),)
plt.ylim((-10, 10))
plt.plot(wave, flux)
plt.show()
```

```python
from scipy.interpolate import interp1d
import pylab as pl
```

```python
pl.figure(num=3, figsize=(18, 3),)
wave=data[1][0]
flux=data[1][1]
f1=interp1d(wave,flux,kind='linear')
```

```python
pl.figure(num=3, figsize=(18, 5),)
pl.ylim((-10,10))
wave=data[1][0]
flux=data[1][1]
f1=interp1d(wave,flux,kind='linear')
waveint=np.linspace(data[1][0][0],data[1][0][-1],2048)
pl.plot(waveint,f1(waveint))
pl.legend(loc="best")
pl.show()
```
